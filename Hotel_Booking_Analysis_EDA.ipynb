## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Create a copy to preserve the original data
df_clean = df.copy()
````

````python
# 1. droping duplicate values using drop_duplicates method
df_clean.drop_duplicates(inplace = True)
````

As there are 31994 duplicate rows in the dataset, so we are simply dropping all the duplicate rows from our dataset

````python
# Columns having missing values.
df_clean.isnull().sum().sort_values(ascending = False)[:5] # checking top 5 columns having NaN values
````

````python
# 2. droping 'agent' and 'company' columns from dataset using 'drop()' method
df_clean = df_clean.drop(columns = ['agent', 'company'])
````

````python
df_clean['children'].unique()
````

````python
# checking value count of 'children' column
df_clean['children'].value_counts()
````

This column 'children' has 0 as value which means 0 children were present in group of customers who made that transaction. So, 'nan' values are the missing values due to error of recording data.

````python
# 3. Handling Missing Values (Updated to avoid FutureWarnings)
df_clean['children'] = df_clean['children'].fillna(0)
df_clean['country'] = df_clean['country'].fillna('Unknown')
````

````python
# Checking if all null values are removed
df_clean.isnull().sum().sort_values(ascending = False)[:6]
````

````python
# 4. Removing rows with remaining tiny nulls
df_clean.dropna(subset=['distribution_channel', 'customer_type', 'reservation_status'], inplace=True)

# 5. Handling "Ghost Bookings" (Zero guests)
df_clean = df_clean[~(df_clean['adults'] + df_clean['children'] + df_clean['babies'] == 0)]

# 6. Data Type Conversion
df_clean['reservation_status_date'] = pd.to_datetime(df_clean['reservation_status_date'])
df_clean['children'] = df_clean['children'].astype(int)

# 7. Feature Engineering
df_clean['total_stay'] = df_clean['stays_in_weekend_nights'] + df_clean['stays_in_week_nights']
df_clean['total_people'] = df_clean['adults'] + df_clean['children'] + df_clean['babies']

# 8. Cleaning 'Undefined' values (Updated to avoid FutureWarnings)
df_clean['meal'] = df_clean['meal'].replace('Undefined', 'SC')

print(f"Data Wrangling Complete. New Shape: {df_clean.shape}")
````

````python
# 9. we can remove this permanently from our database

df_clean.drop(df_clean[df_clean['adr'] < 0].index, axis = 0, inplace= True)
````

````python
# checking if adr < 0 deleted
df_clean[df_clean['is_canceled'] == 0]['adr'].describe()
````

### What all manipulations have you done and insights you found?

Handling Missing Values (Modern Imputation): I replaced null values in children, agent, and company with 0 and in country with 'Unknown'. I specifically transitioned to the .fillna() assignment method to ensure the code is future-compatible and avoids FutureWarnings.

Removal of "Ghost Bookings": I filtered out rows where the sum of adults, children, and babies was zero. These entries represent invalid bookings with no guests, which would have skewed the Average Daily Rate (ADR) and occupancy metrics.

Deduplication: I removed over 14,000 duplicate rows. This was a critical step to ensure that our statistical analysis of cancellation rates and guest behavior is based on unique transactions rather than repeated system logs.

Outlier Treatment: "I identified and removed a row with a negative adr value. Since the Average Daily Rate cannot be negative in a standard business context, this row was treated as a data entry error. Removing it ensures our revenue and pricing analysis remains accurate."

Type Casting for Optimization: I converted reservation_status_date to a datetime object and changed columns like children, agent, and company from floats to integers. This makes the data more memory-efficient and prevents visual "noise" (like "2.0 children") in charts.

Feature Engineering: I created two new columns:

total_stay: Combining weekend and weekday nights to see the full length of a guest's visit.

total_people: Combining all guest categories to understand the actual load on hotel resources.

Categorical Cleaning: I merged 'Undefined' meal types into the 'SC' (Self-Catering) category to simplify the distribution and make the visualizations more meaningful.
